```python
import numpy as np
import pandas as pd
```

```python
def step_gradient(points,learning_rate,m):
    M=len(points)
    m_slope=list([0 for i in range(len(points[0])-1)])
    for i in range(M):
        y=points[i,len(points[0])-2]
        x_total=0;
        q=0
        for k in range(len(points[0])):
            if(k==len(points[0])-2):
                continue
            x_total+=m[q]*points[i,k]
            q=q+1
        l=0   
        for j in range(len(points[0])):
            if (j==len(points[0])-2):
                continue
            m_slope[l]+=(-2/M)*((y-x_total)*points[i,j])
            l=l+1
    new_m=list([0 for j in range(len(points[0])-1)])
    a=0
    for i in range(len(m)):
        new_m[i]=m[i]-learning_rate*m_slope[i]
    return new_m    
```

```python
def gd(points,learning_rate,num_iteration):
    m=list([0 for j in range(len(points[0])-1)])
    for i in range(num_iteration):
        m=step_gradient(points,learning_rate,m)
        print(i," cost : ",cost(points,m))
    return m    
        
```

```python
def run():
    d=np.genfromtxt("/home/spari/Music/ML/boston_dataset/0000000000002417_training_boston_x_y_train.csv",delimiter=",")
    b=np.ones((len(d),1))
    c=np.hstack((d,b))
    df=pd.DataFrame(c)
    x2=df.values        
    learning_rate=.1
    num_iteration=898
    m=gd(x2,learning_rate,num_iteration)
    return m    
```

```python
def cost(points,m):
    total_cost=0
    M=len(points)
    total_cost=0
    for i in range(M):
        y=points[i,len(points[0])-2]
        x_total=0
        q=0
        for k in range(len(points[0])):
            if(k==len(points[0])-2):
                continue
            x_total+=m[q]*points[i,k]
            q=q+1
        total_cost+=(1/M)*((y-x_total)**2)    
    return total_cost    
```

```python
def predict(p,m):
    y_pred=list([0 for j in range(len(p))])
    k=0
    for i in range(len(p)):
        for j in range(len(m)):
            y_pred[k]+=m[j]*p[i,j]
        k=k+1    
    return y_pred
```
