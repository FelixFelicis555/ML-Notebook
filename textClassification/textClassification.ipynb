```python
from nltk.corpus import stopwords
import os
import re
import pandas as pd
import numpy as np
```

```python
stop_words = set(stopwords.words('english'))
block_wrds = ['last','better','never','every','even','two','good','used','first','need','going','must','really','might','well','without','made','give','look','try','far','less','seem','new','make','many','way','since','using','take','help','thanks','send','free','may','see','much','want','find','would','one','like','get','use','also','could','say','us','go','please','said','set','got','sure','come','lot','seems','able','anything','put']
not "like" in block_wrds
```

strip the documents and take into dictionary after the header only!

```python
dictionary = {}
type(dictionary)
count=0
for file in os.listdir("mini_newsgroups"):
    for files in os.listdir("mini_newsgroups/"+file):
        #print(file,files)
        f = open("mini_newsgroups/"+file+"/"+files,'r',errors='ignore')
        message = f.read()
        message = message.split()
        k =0
        for i in message:
            if(i=='Lines:'):
                k = 1
                count +=1
            if(k== 1):
                if(i.isalpha() == True and len(i) > 1):
                    if not i.lower() in stop_words:
                        if not i.lower() in block_wrds:
                            if(i.lower() in dictionary.keys()):
                                dictionary[i.lower()] = dictionary[i.lower()] +1
                            else:
                                dictionary[i.lower()] = 1


        f.close()
```

```python
dictionary
```

```python
import operator
sorted_vocab = sorted(dictionary.items(), key= operator.itemgetter(1), reverse= True)
sorted_vocab
```

```python
top_val = []
sorted_vocab[1000][1]
size = len(sorted_vocab)
for i in range(size):
    if(sorted_vocab[1000][1] <= sorted_vocab[i][1]):
        top_val.append(sorted_vocab[i][0])
```

```python
top_val[0:100]
```

```python
df = pd.DataFrame(columns = top_val)
df.columns
```

```python
df = pd.DataFrame(columns = top_val)
df.columns
count=0
for file in os.listdir("mini_newsgroups"):
    for files in os.listdir("mini_newsgroups/"+file):
        count=count+1
        #print(file,files)
        df.loc[len(df)] = np.zeros(len(top_val))
        f = open("mini_newsgroups/"+file+"/"+files,'r',errors='ignore')
        message = f.read()
        message = message.split()
        k =0
        for i in message:
            if(i in df.columns):
                df[i][len(df)-1] += 1
        f.close()
count
```

```python
y=[]
i=0
for file in os.listdir("mini_newsgroups"):
    for files in os.listdir("mini_newsgroups/"+file):
        #print(file,files)
        y.append(i)
    i=i+1
```

```python
y = np.array(y)
y.shape,df.shape
x = df.values
```

```python
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)

train_score = clf.score(x_train, y_train)
test_score = clf.score(x_test, y_test)

train_score, test_score
```
