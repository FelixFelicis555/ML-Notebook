```python
import numpy as np
from sklearn import preprocessing
```

```python
def step_gradient(points, learning_rate , m ):
    m_slope=np.zeros(points.shape[1])
    new_m=np.zeros(points.shape[1])
    M= points.shape[0] 
    N= points.shape[1]
    x_unscaled = points[:,0:N-1]
    y_train=points[:,N-1]

    x_scaled= preprocessing.scale(x_unscaled)
 
    
    
    for j in range(N):
        for i in range(M) :
            y= y_train[i] #y_train value for each row
            x=x_scaled[i]       #1st row
            x=np.append(x,1)         #appendind 1 in the end of the row
            #print("x ",x, "y", y )
            #print("x.shape",x.shape , "y shape ",y.shape ,"m shap",m.shape)
            mul=(m*x).sum()   #m1x1* m2*x2.....
            m_slope[j] += (-2/M)* (y - mul) *x[j]
            
        new_m[j] = m[j] -learning_rate*m_slope[j]
    
    
    return new_m 
    #return new_m , new_c
        
```

```python
def gd(points,learning_rate,num_iterations):
    m=np.zeros(points.shape[1])
    for i in range(num_iterations):
        
        m= step_gradient(points, learning_rate , m )
        print(i, "cost ", cost(points,m ))
    return m 

```

```python
def cost(points, m ):
    total_cost =0
    M= points.shape[0] 
    N= points.shape[1]
    for i in range(M) :
            y=points[i][N-1]
            x=points[i]
            x[N-1]=1
            mul=(m*x).sum()
            total_cost += (1/M)*((y - mul)**2)
            
       
    return total_cost


    
```

```python
def predict(data_test,m):
    M=data_test.shape[0]
    N=data_test.shape[1]
    y_pred=np.zeros(M)
    #c=m[N]
    #m=m[0:N]
    for i in range(M):
        x_test=data_test[i]
        
        x_test=np.append(x_test,1)
        
        #print("x",x, "m" , m)
        y_pred[i]= (x_test*m).sum()
    return y_pred
```

```python
def run():
    data =np.genfromtxt("training_ccpp_x_y_train.csv",delimiter =",")
    learning_rate=0.1
    num_iterations=300
    m =gd(data,learning_rate,num_iterations)
    print(m)
    data_test=np.genfromtxt("test_ccpp_x_test.csv",delimiter =",")
    y_pred=predict(data_test,m)
    print(y_pred)
    np.savetxt('pakhiccpp.csv',y_pred ,delimiter=",",fmt="%.5f") 
```
