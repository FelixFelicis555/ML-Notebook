```python
N=50
a = np.random.rand(N)
b = np.random.rand(N)
colors = np.random.rand(N)
area=  np.pi * (15 * np.random.rand(N))**2
plt.scatter(a,b,s=area,c=colors,alpha=0.5)
plt.show()
```
